```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../data/customer_pca.csv')
data.head()
```

```python
data_x = data.drop('label', axis = 1)
data_y = data['label']
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca.fit(data_x)
data_pca = pca.transform(data_x)
data_pca
```

```python
data_pca = pd.DataFrame(data_pca, columns = ['PC1', 'PC2'])
data_pca = data_pca.join(data_y)
data_pca.head()
```

```python
sns.scatterplot(x = 'PC1', y = 'PC2', data = data_pca, hue = 'label', 
               palette = 'rainbow')
```

```python
df_comp = pd.DataFrame(pca.components_, columns = data_x.columns)
```

```python
sns.heatmap(df_comp, cmap = 'coolwarm')
```

```python
data = pd.read_csv('../data/anonymous.csv')
data.head()
```

```python
data['class'].mean()
```

```python
data.isna().sum().sum()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data.drop('class', axis = 1), 
    data['class'], test_size= 0.2, random_state = 100)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_1 = RandomForestClassifier(random_state = 100)
```

```python
import time
start_time = time.time()
model_1.fit(x_train_scaled, y_train)
print(time.time() - start_time)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

pred_1 = model_1.predict(x_test_scaled)
accuracy_score(y_test, pred_1)
```

```python
data['class'].mean()
```

```python
proba_1 = model_1.predict(x_test_scaled)
```

```python
roc_auc_score(y_test, proba_1)
```

```python
pca = PCA(n_components = 2)
pca.fit(x_train_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
var_ratio = []

for i in range(100, 550, 50):
    pcas = PCA(n_components = i)
    pca.fit_transform(x_train_scaled)
    ratio = pca.explained_variance_ratio_.sum()
    var_ratio.append(ratio)

sns.lineplot(x = range(100, 550, 50), y = var_ratio)
```

```python
pca = PCA(n_components = 400, random_state = 100)
pca.fit(x_train_scaled)
x_train_scaled_pca = pca.transform(x_train_scaled)
x_test_scaled_pca = pca.transform(x_test_scaled)
```

```python
model_2 = RandomForestClassifier(random_state = 100)
start_time = time.time()
model_2.fit(x_train_scaled_pca, y_train)
print(time.time() - start_time)
```

```python
pred_2 = model_2.predict(x_test_scaled_pca)
accuracy_score(y_test, pred_2)
```

```python
proba_2 = model_2.predict_proba(x_test_scaled_pca)
roc_auc_score(y_test, proba_2[:, 1])
```
