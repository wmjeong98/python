```python
tips = sns.load_dataset("tips")
tips
```

```python
fig1 = plt.figure()

ax1 = fig1.add_subplot(1,1,1)
ax1.hist(data = tips, x = 'total_bill', bins = 10)

ax1.set_title('Histogram of Total Bill')
ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Frequency')

plt.show()
```

```python
scatter_plot = plt.figure()
ax2 = scatter_plot.add_subplot(1,1,1)

ax2.scatter(tips['total_bill'], tips['tip'])
ax2.set_title('Scatterplot of Total Bill vs Tip')
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')

plt.show()
```

```python
boxplot = plt.figure() # Shift + Tab :         
ax3 = boxplot.add_subplot(1,1,1)

ax3.boxplot(
    x = [
        tips[tips['sex'] == 'Female']['tip'],
        tips[tips['sex'] == 'Male']['tip']
    ],
    tick_labels = ['Female', 'Male'] #labels Matplotlib 3.9  tick_labels   labels    
)

ax3.set_xlabel('Sex')
ax3.set_ylabel('Tip')
ax3.set_title('Boxplot of Tips by Sex')

plt.show()
```

```python
colors = {"Female" : "#f1a340", "Male" : "#998ec3"}

scatter_plot = plt.figure()
ax4 = scatter_plot.add_subplot(1,1,1)

ax4.scatter(data = tips,
            x = 'total_bill',
            y = 'tip',
            s = tips['size']**2*10,
            c = tips['sex'].map(colors),
            alpha = 0.5
)

ax4.set_title('Colored by Sex and Sized by Size')
ax4.set_xlabel('Total Bill')
ax4.set_ylabel('Tip')

scatter_plot.suptitle('Total Bill vs Tip')

plt.show()
```

```python
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

fig = plt.figure()
for idx,  style in enumerate(seaborn_styles):
    plot_position = idx + 1;
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(data = tips, x = 'time', y = 'total_bill', ax = ax)
        violin.set_title(style)
plt.show()
```

```python
import pandas as pd

contexts = pd.DataFrame({
    "paper" : sns.plotting_context("paper"),
    "notebook" : sns.plotting_context("notebook"),
    "talk" : sns.plotting_context("talk"),
    "poster" : sns.plotting_context("poster")
})
print(contexts)
```

```python
context_styles = contexts.columns

fig = plt.figure()

for idx, context in enumerate(context_styles):
    plot_position = idx + 1
    with sns.plotting_context(context):
        ax = fig.add_subplot(2, 2, plot_position)
        violin = sns.violinplot(data = tips, x = 'time', y = 'total_bill', ax = ax)
        violin.set_title(context)

fig.set_tight_layout(True)
plt.show()
```
